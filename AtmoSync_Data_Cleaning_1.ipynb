## 5. Standardize Column Names

Column names are normalized to a consistent `snake_case` format (stripped of
whitespace, lower-cased, spaces replaced with underscores) so every later
reference to a column is predictable.

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.columns.tolist()
```

## 6. Fix the Timestamp Column

The `timestamp` column is read in as plain text by default. Converting it to a
real `datetime` type (and renaming it to the clearer `date_time`) is what makes
time-based sorting, resampling, and plotting possible later on.

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
```

```python
# Confirm the conversion didn't silently fail on any row
df['timestamp'].isnull().sum()
```

```python
df.rename(columns={'timestamp': 'date_time'}, inplace=True)
```

```python
df[['container_id', 'date_time']].head()
```

## 7. Clean Text Columns

Trailing/leading whitespace in text fields is a common source of "phantom"
categories (e.g. `"Avocado"` vs `"Avocado "` being treated as different groups).
Stripping every text column up front avoids that class of bug.

```python
text_cols = df.select_dtypes(include='object').columns.tolist()
for col in text_cols:
    df[col] = df[col].str.strip()
text_cols
```

## 8. Correct Data Types

* Repeated-category columns (commodity, ports, markets, risk status, recommended
  action) are cast to `category` — this is more memory-efficient and makes
  grouping/plotting cleaner.
* `cooling_unit_faulty` is a true boolean flag and is cast to `bool` explicitly.

```python
category_cols = [
    "commodity", "origin_port", "primary_market",
    "secondary_market", "risk_status", "recommended_action",
]
for col in category_cols:
    df[col] = df[col].astype("category")

df["cooling_unit_faulty"] = df["cooling_unit_faulty"].astype("bool")
df.dtypes
```

## 9. Sort Records Logically

Sensor readings should be ordered by container and then by time — this makes the
data usable for time-series analysis per container (e.g. tracking one shipment's
temperature over its whole transit).

```python
df = df.sort_values(["container_id", "date_time"]).reset_index(drop=True)
df.head()
```

## 10. Logical Consistency Checks

Beyond missing values and duplicates, it's worth checking that the data makes
physical sense: ideal temperature ranges should be valid (min < max), and
quantities/prices/transit-time shouldn't be negative.

```python
print("Rows where ideal_temp_min >= ideal_temp_max:",
      (df["ideal_temp_min_c"] >= df["ideal_temp_max_c"]).sum())
print("Rows with negative hours_in_transit:", (df["hours_in_transit"] < 0).sum())
print("Rows with negative quantity_kg:", (df["quantity_kg"] < 0).sum())
print("Rows with negative primary price:", (df["primary_price_per_kg_usd"] < 0).sum())
print("Rows with negative secondary price:", (df["secondary_price_per_kg_usd"] < 0).sum())
```

All checks pass — the dataset has no logically impossible values, so no rows
need to be dropped or corrected on this basis.

## 11. Engineer a Data Quality Flag

A `data_quality_flag` column is added to make it easy to filter out readings with
significant temperature drift (deviation from the ideal range greater than 3°C)
in later analysis, without deleting any raw data.

```python
df["data_quality_flag"] = "Normal Reading"
df.loc[df["temp_deviation_c"] > 3, "data_quality_flag"] = "Significant Temperature Drift"
df["data_quality_flag"].value_counts()
```
